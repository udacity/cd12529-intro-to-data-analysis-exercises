## Code

```python
import numpy as np
import pandas as pd


# Generate 200 random data points along 3 dimensions
x, y, scale = np.random.randn(3, 200)
```

```python
df = pd.DataFrame({"x": x, "y": y, "scale": np.abs(scale)*400})
df.head()
```

## Visualizations

```python
df.plot.scatter(x="x", y="y", s="scale", c="scale", colormap='viridis', figsize=(12, 8));
```
